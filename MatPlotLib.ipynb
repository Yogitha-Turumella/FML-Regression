# **MATPLOTLIB**


*   John Hunter introduced the Matplotlib platform in 2002.
*   It comprises various plots like bar charts,line graphs,scatter plots,each providing a different way to display data.



*   Matplotlib is a plotting library for the python programming language and its numerical mathematics extension Numpy.
*   It is used for creating static, animated, and open-source visualization library in python.

*   The six most commonly used plots come under matplotlib.These are:
*   Line Plot
*   Scatter Plot
*   Bar Plot
*   Histogram Plot
*   Box Plot
*   Pie Plot



















```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

## **Line Plot**



*   It is drawn by joining straight lines.
*   In matplotlib , the plot() function represents this.





```python
x=np.arange(0,10,2)
y=x**2
print(x,y)
```

```python
plt.plot(x,y,'r-.')
```

```python
plt.plot(x,y,'r-.')    # r means red color
plt.xlabel('X-axis')   # instead of '-.' we can also use 'ls=dashdot'
plt.ylabel('Y-axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
plt.plot(x,y,ls='dotted')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph in 2D')
```

```python
plt.plot(x,y,ls="none")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph in 2D')
```

```python
plt.plot(x,y,marker=".",c="r")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Graph in 2D')
```

```python
plt.plot(x,y,c='r',lw=5,ls='dotted') # c means color # lw means line width
```

```python
x = np.arange(0,10,2)
y1 = x**2
y2 = x**3
y3 = x**4
print(x,y1)
print(x,y2)
print(x,y3) # shows multiple lines in a single graph
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
```

```python
# subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g--')
plt.subplot(2,2,3)
plt.plot(x,y,'o')
plt.subplot(2,2,4)
plt.plot(x,y,'bo')
```

# **Scatter Plot**

*   Scatter plot uses dots to represent values for two different numeric variables.
*   It is used to observe relationships between variables.
*   The function used for this is scatter().






```python
x = [2,4,6,8,10,12]
y = [50,70,20,80,90,40]
plt.scatter(x,y,c='maroon') # it relates x with y (representation in the form of plot)
```

```python
x = [2,4,6,8,10,12]
y = [50,70,20,80,90,40]
c =["orange","yellow","purple","black","green","maroon"]
plt.scatter(x,y,c=c) # it relates x with y (representation in the form of plot)
```

```python
x = [2,4,6,8,10,12]
y = [50,70,20,80,90,40]
#c =["orange","yellow","purple","black","green","maroon"]
size=[100,200,50,150,300,250]
plt.scatter(x,y,color="maroon",s=size,alpha=0.5) # alpha is used to add transparency to a color
plt.colorbar()                                                #by default aplha=1
```

```python
x = [2,4,6,8,10,12]
y = [50,70,20,80,90,40]
#c =["orange","yellow","purple","black","green","maroon"]
size=[100,200,50,150,300,250]
plt.scatter(x,y,c=c,s=size)
x = [2,4,6,8,10,12]
y = [25,65,95,75,55,35]
#c =["orange","yellow","purple","black","green","maroon"]
size=[100,200,50,150,300,250]
plt.scatter(x,y,c=c,s=size)
plt.legend()
plt.title("scatter plot",color="red")
plt.colorbar()
```

# **Bar Plot**

*   Bar Plot is a type of data visualization used to represent data in form of rectangular bars.
*   The function used for Bar plot is bar().
*   The height of each bar represents the value of the data point , and the width of each bar represents the category of the data.


```python
x = ['H','E','M','A']
y = [30,50,70,90]
plt.bar(x,y)
```

```python
x = ['H','E','M','A']
y = [30,50,70,90]
plt.bar(x,y,color="maroon")
```

```python
x = ['H','E','M','A'] #vertical Bar Representation
y = [30,50,70,90]
plt.bar(x,y,color="orange",width=0.5) #by default width will takes 0.8
```

```python
x = ['H','E','M','A'] #horizontal Bar Representation
y = [30,50,70,90]
plt.barh(x,y,color="pink",height=0.5) #by default height will takes 0.8
```

```python
x = ['H','E','M','A'] #horizontal Bar Representation
y = [30,50,70,90]
c=["maroon","black","pink","yellow"]
plt.barh(x,y,color=c,height=0.5) #by default height will takes 0.8
plt.show()
```

```python
x=["2020","2021","2022","2023"] #vertical Bar Representation
y= [280,350,470,590]
c=["green","skyblue","pink","yellow"]
plt.bar(x,y,color=c,width=0.3) #b default width will takes 0.8
plt.xlabel("Year")
plt.ylabel("No of Students Passed")
plt.title("Placements Data",color="purple")
```

# **Histogram**

*   Histogram is a graph showing frequency distributions.
*   It is a graph showing the number of observations within each given interval.
*  It is an accurate representation of the distribution of numerical data.






```python
marks=[90,95,20,35,70,60,65,30,55]
grade_intervals=[0,50,80,100]
plt.title("Student Grade")
plt.hist(marks,grade_intervals,histtype="bar",rwidth=0.7) #reduce width
plt.xticks([0,50,80,100])
plt.xlabel("Percentage")
plt.ylabel("No.of Students")

```

```python
marks=[90,95,20,35,70,60,65,30,55]
grade_intervals=[0,50,80,100]
plt.title("Student Grade")
plt.hist(marks,grade_intervals,histtype="step",rwidth=0.7) #reduce width
plt.xticks([0,50,80,100])
plt.xlabel("Percentage")
plt.ylabel("No.of Students")
```

```python
marks=[90,95,20,35,70,60,65,30,55]
grade_intervals=[0,50,80,100]
plt.title("Student Grade")
plt.hist(marks,grade_intervals,histtype="stepfilled",rwidth=0.7,facecolor="r") #reduce width
plt.xticks([0,50,80,100])
plt.xlabel("Percentage")
plt.ylabel("No.of Students")
```

```python
marks=[90,95,20,35,70,60,65,30,55]
grade_intervals=[0,50,80,100]
plt.title("Student Grade")
plt.hist(marks,grade_intervals,histtype="bar",rwidth=0.7,facecolor="g") #reduce width
plt.xticks([0,50,80,100])
plt.xlabel("Percentage")
plt.ylabel("No.of Students")
```

# **Box Plot**

*   Essentially a Box Plot describes a few statistical information of a given data using box.it can be representes both horizontally and vertically
*   A quartile is a statistical phrase for dividing observations into four predetermined intervals based on data values.



```python
list1=[11,2,4,6,7,8,3]
fig=plt.figure(figsize=(10,8))
plt.boxplot(list1)
```

```python
list2=[1,2,2,2,3,5,6]
fig=plt.figure(figsize=(10,7))
plt.boxplot(list2)
```

```python
list1=[11,4,6,8,6,9,3]
fig=plt.figure(figsize=(5,3))
plt.boxplot(list1,vert=False)
```

```python
list1=[11,4,6,8,6,9,3]
fig=plt.figure(figsize=(5,4))
plt.boxplot(list1,widths=0.6)
```

```python
list1=[11,4,6,8,6,9,3]
fig=plt.figure(figsize=(5,4))
plt.boxplot(list1,labels=["samrin"],widths=0.5)
```

```python
list1=[11,4,6,8,6,9,3]
fig=plt.figure(figsize=(5,4))
plt.boxplot(list1,labels=["samrin"],widths=0.3,patch_artist = True)
```

# **Pie Chart**

*   A Pie Chart is a circular statistical plot that can display only one series of data.
*   The area of the chart is the total percentage of the given data.
*   The area of slices of the pie represents the percentage of the parts of the data.
*   The slices of pie are called wedges.





```python
Student_performance=['Excellent','Good','Average','poor']
Student_values=[35,30,20,15]
plt.pie(Student_values)
```

```python
student_performance=['Excellent','Good','Average','Poor']
student_values=[35,30,20,15]
plt.pie(student_values , labels =student_performance,startangle=0,explode=[0.3,0,0,0])
```

```python
student_performance=['Excellent','Good','Average','Poor']
student_values=[35,30,20,15]
plt.pie(student_values , labels =student_performance,startangle=0,explode=[0.2,0.2,0.2,0.2])
```

```python
student_performance=['Excellent','Good','Average','Poor']
student_values=[35,30,20,15]
plt.pie(student_values , labels =student_performance,startangle=0,explode=[0.2,0.2,0.2,0.2],
colors=["pink","yellow","purple","skyblue"])
```

```python
student_performance=['Excellent','Good','Average','Poor']
student_values=[35,30,20,15]
plt.pie(student_values , labels =student_performance,startangle=0,explode=[0.2,0.2,0.2,0.2],colors=["pink","yellow","purple","skyblue"])
plt.legend(title ="student performance")
```
